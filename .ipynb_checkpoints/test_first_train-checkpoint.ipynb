```python
class CustomConfig(Config):
    """Configuration for testing on the custom  dataset.
    Derives from the base Config class and overrides some values.
    """
    # Give the configuration a recognizable name
    NAME = "object"

    IMAGES_PER_GPU = 1

    NUM_CLASSES = 1 + 2  # Background + Lion and Tigre

    # Number of training steps per epoch
    STEPS_PER_EPOCH = 1

    # Skip detections with < 90% confidence
    DETECTION_MIN_CONFIDENCE = 0.5
    
```

```python
import os
import json
import numpy as np
import skimage.draw
import matplotlib.pyplot as plt
from mrcnn import utils

# Mode d'inspection du modèle
TEST_MODE = "inference"

# Chemins racine de votre projet et de votre dataset
ROOT_DIR = r"C:\Users\bouba\Nouvelle_vie\Mask_RCNN"
DATASET_DIR = os.path.join(ROOT_DIR, "dataset")

# Classe CustomDataset mise à jour pour prendre en compte les chemins sans sous-dossier d'image
class CustomDataset(utils.Dataset):
    def load_custom(self, json_dir, subset):
        """Charge les données personnalisées avec un dossier 'json' contenant les JSON et les images.
        `json_dir` est le dossier des fichiers JSON pour le subset ('train' ou 'val').
        """
        # Ajouter les classes
        self.add_class("custom", 1, "lion")
        self.add_class("custom", 2, "tigre")

        # Parcourir les fichiers JSON dans le dossier spécifique
        for filename in os.listdir(json_dir):
            if not filename.endswith(".json"):
                continue

            json_path = os.path.join(json_dir, filename)
            try:
                # Charger le fichier JSON
                with open(json_path, 'r', encoding='utf-8') as json_file:
                    data = json.load(json_file)
            except Exception as e:
                print(f"Erreur de chargement du JSON pour {filename}: {e}")
                continue

            # Vérifier les informations nécessaires directement dans `data`
            if 'shapes' not in data or 'imageHeight' not in data or 'imageWidth' not in data or 'imagePath' not in data:
                print(f"Le fichier {filename} manque de certaines informations nécessaires.")
                continue

            polygons = []
            num_ids = []
            for shape in data['shapes']:
                label = shape.get('label')
                if label == "lion":
                    num_ids.append(1)
                elif label == "tigre":
                    num_ids.append(2)
                else:
                    continue
                polygons.append(shape['points'])

            height = data['imageHeight']
            width = data['imageWidth']
            image_path = os.path.join(json_dir, data['imagePath'])

            # Vérifier l'existence de l'image
            if not os.path.exists(image_path):
                print(f"Image non trouvée pour {filename} ({image_path}), ignorée.")
                continue

            # Ajouter l'image au dataset
            self.add_image(
                source="custom",
                image_id=filename,
                path=image_path,
                width=width,
                height=height,
                polygons=polygons,
                num_ids=num_ids
            )

# Fonction pour obtenir des axes pour la visualisation
def get_ax(rows=1, cols=1, size=16):
    _, ax = plt.subplots(rows, cols, figsize=(size*cols, size*rows))
    return ax

# Charger le dataset de validation
CUSTOM_VAL_DIR = os.path.join(DATASET_DIR, "val", "json")  # Dossier contenant les JSON et les images pour la validation
dataset = CustomDataset()
dataset.load_custom(CUSTOM_VAL_DIR, "val")  # Charger les données de validation
dataset.prepare()
# print("Images: {}\nClasses: {}".format(len(dataset.image_ids), dataset.class_names))
```
